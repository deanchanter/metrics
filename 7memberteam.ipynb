Simulation Setup 
Run: https://github.com/deanchanter/metrics/blob/master/WIPStudy_Average.py
Runs = 1000000
WIPLimits = [3,5,7,11,13,21]

7 "pickers" (People) have a 50/50 problaby of completing a work item each day. 
If the team has not reached the WIP Limit work is started if work can not complete.
Items are completed in First in First Out Order.
The Simulation burns down 100 items 

```python
import pandas as pd
#import data
time_to_completedf = pd.read_csv("Documents/git/metrics/7 member team 3-21/time.csv", index_col=0)
cycletimedf = pd.read_csv("Documents/git/Metrics/7 member team 3-21/ct.csv",index_col=0)  
throughputdf = pd.read_csv("Documents/git/Metrics/7 member team 3-21/tp.csv",index_col=0)
max_agedf = pd.read_csv("Documents/git/Metrics/7 member team 3-21/ctmax.csv",index_col=0)
average_wipdf = pd.read_csv("Documents/git/Metrics/7 member team 3-21/wip.csv",index_col=0)
```

Plot Histograms of Average Cycle Time, Throughtput, Work In Progress

```python
ct = cycletimedf.plot.hist(alpha=.7, bins=25)
ct.set(title="Cycle Time",xlabel="Average Cycle Time (Days)")

tp = throughputdf.plot.hist(alpha=.7, bins=20)
tp.set(title= "Throughput", xlabel=" Average # of Items per Day")

rw = average_wipdf.plot.hist(alpha=.7, bins=30)
rw.set(title="WIP", xlabel= "Average # of Items in Progress")
```

Little's Law tracks even in this "random" system. Notice how how average WIP is allowed to move towards random (no limit) cycle time follows the same pattern.  Even when allowed to work on up to 21 items at a time the fixed set of "resources" only accomplish on average .5 item per day more

So how much faster is that really?
