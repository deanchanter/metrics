```python
import pandas as pd
#import data
time_to_completedf = pd.read_csv("Documents/git/Metrics/7 member team 3-21/time.csv", index_col=0)
cycletimedf = pd.read_csv("Documents/git/Metrics/7 member team 3-21/ct.csv",index_col=0)  
throughputdf = pd.read_csv("Documents/git/Metrics/7 member team 3-21/tp.csv",index_col=0)
max_agedf = pd.read_csv("Documents/git/Metrics/7 member team 3-21/ctmax.csv",index_col=0)
average_wipdf = pd.read_csv("Documents/git/Metrics/7 member team 3-21/wip.csv",index_col=0)
```

```python
ct = cycletimedf.plot.hist(alpha=.7, bins=25)
ct.set(title="Cycle Time",xlabel="Average Cycle Time (Days)")

tp = throughputdf.plot.hist(alpha=.7, bins=20)
tp.set(title= "Throughput", xlabel=" Average # of Items per Day")

rw = average_wipdf.plot.hist(alpha=.7, bins=30)
rw.set(title="WIP", xlabel= "Average # of Items in Progress")
```
